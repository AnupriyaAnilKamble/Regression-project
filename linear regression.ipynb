```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_dataset_linearregression.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df['YearsExperience']
X
```

```python
y = df['Salary']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
type(X_train)
```

```python
X_tr = pd.DataFrame(X_train)
y_tr = pd.DataFrame(y_train)
```

```python
type(X_tr)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_tr, y_tr)
```

```python
regressor.intercept_
```

```python
X_test
```

```python
regressor.predict(pd.DataFrame(X_test))
```

```python
regressor.predict(pd.DataFrame(X_train))
```

```python
y_train
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(pd.DataFrame(X_train)), color='blue')
plt.title('Salary Vs Experience graph')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, regressor.predict(pd.DataFrame(X_test)), color='blue')
plt.title('Salary Vs Experience graph')
plt.xlabel('Experience')
plt.ylabel('Salary')
```
